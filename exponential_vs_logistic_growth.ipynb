```python
from tkinter import font
import numpy as np

import matplotlib.pyplot as plt

# Parameters for the models
time = np.linspace(0, 10, 500)
initial_population = 10
growth_rate = 0.5
carrying_capacity = 200

# Exponential growth model
exponential_growth = initial_population * np.exp(growth_rate * time)

# Logistic growth model
logistic_growth = carrying_capacity / (1 + (carrying_capacity - initial_population) / initial_population * np.exp(-growth_rate * time))

# Plotting the models
plt.figure(figsize=(10, 6))
plt.plot(time, exponential_growth, label="Exponential Growth")
plt.plot(time, logistic_growth, label="Logistic Growth")
plt.title(r"Exponential Growth vs Logistic Growth, $P_0=1$, $r=0.5$, $K=200$", fontsize=18)
plt.xlabel("Time", fontsize=16)
plt.ylabel("Population", fontsize=16)
plt.legend(fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid()
plt.tight_layout()
plt.savefig("LatexReport/Images/Plots/Created/exponential_vs_logistic_growth.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# MONOD Equation 
x = np.linspace(0, 500, 10000)
carrying_capacity = 10
v = 1
monod = (x * v) / (x + carrying_capacity)
plt.plot(x, monod, label="Monod Equation, K=10, v=1")
plt.axvline(x=10, color='red', linestyle='--', label='Resources=10')
plt.axhline(y=0.5, color='green', linestyle='--', label='y=0.5')
plt.title("Monod Equation", fontsize=18)
plt.xlabel("Substrate Concentration", fontsize=16)
plt.ylabel("Reaction Rate", fontsize=16)
plt.legend(fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('LatexReport/Images/Plots/Created/monod_equation.png', dpi=300, bbox_inches='tight')
```
