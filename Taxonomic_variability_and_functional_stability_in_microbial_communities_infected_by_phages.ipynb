```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def plotdata(t, y):
    plt.plot(t, y)
    plt.xlabel('Time')
    plt.ylabel('Growth of Population')
    plt.title('Simple ODE System')
    plt.legend(['Uninfected', 'Infected', 'Phages'])
    plt.show()

def system1(y, t, Yr, ms, H, D, beta, mu, delta, v):
    U, I, P = y
    dUdt = Yr / ms * U * H - D * U - beta * P * U
    dIdt = beta * P  * U - D * I - mu * I
    dPdt = v * mu * I - beta * P * (U + I) -(D + delta) * P
    print(dUdt, dIdt, dPdt)
    return [dUdt, dIdt, dPdt]
```

```python
Yr = 15.6
ms = 440
H = 0.002
D = 0.1
beta = 0.000000003 * 0.1
mu = 4
delta = 0.5
v = 0.4
U = 100
I = 0
P = 10

# Combine initial conditions into a single array
y0 = [U, I, P]

# Time points where solution is computed
t1 = np.linspace(0, 20, 300)

y1 = odeint(system1, y0, t1, args=(Yr, ms, H, D, beta, mu, delta, v))
plotdata(t1, y1)
```
