```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def generalized_lotka_volterra2(t, B, r, alpha, m):
    dBdt = B * (r + alpha @ B - m)
    return dBdt

B0 = np.array([100, 200])
r = np.array([0.5, 0.13])
alpha = np.array([[-0.49, 0.47], [0.5, -0.48]])
m = np.array([0.5, 0.3])
t = np.linspace(0, 20, 1000)

B_solution = solve_ivp(generalized_lotka_volterra2, (0, 20), B0, t_eval=t, args=(r, alpha, m))

plt.plot(B_solution.t, B_solution.y[0], 'r', label='Bacteria 1')
plt.plot(B_solution.t, B_solution.y[1], 'b', label='Bacteria 2')
plt.legend()
plt.grid()
```
